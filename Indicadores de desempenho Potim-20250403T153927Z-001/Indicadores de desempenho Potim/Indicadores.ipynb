### Vamos comparar os 3 Quadrimestre de 2024 de cada ubs
1 passo: juntar todos os arquivos de cada Quadrimestre em 1 usando o metodo concat:

```python
# Carregar os 3 CSVs adicionando a coluna "Quadrimestre"
df_q1 = pd.read_csv("/content/drive/MyDrive/Aulas Python /Pandas/Indicadores de desempenho Potim/indicador-1Q.csv", encoding='latin1', sep=';')
df_q1["Quadrimestre"] = "1º Quadrimestre"

df_q2 = pd.read_csv("/content/drive/MyDrive/Aulas Python /Pandas/Indicadores de desempenho Potim/indicador-2Q.csv", encoding='latin1', sep=';')
df_q2["Quadrimestre"] = "2º Quadrimestre"

df_q3 = pd.read_csv("/content/drive/MyDrive/Aulas Python /Pandas/Indicadores de desempenho Potim/indicador-3Q.csv", encoding='latin1', sep=';')
df_q3["Quadrimestre"] = "3º Quadrimestre"

# Juntar todos os DataFrames em um só
df_total = pd.concat([df_q1, df_q2, df_q3], ignore_index=True, axis=0, join='outer')
#O atributo axis=0 (padrão) significa que os dados serão empilhados verticalmente, ou seja, as colunas permanecerão as mesmas e novas linhas serão adicionadas. Esse comportamento é utilizado, por exemplo, com pd.concat([df1, df2]) no Pandas.
#O parâmetro join define como lidar com colunas que não são comuns entre os DataFrames:
#join='inner' → Mantém apenas as colunas comuns aos DataFrames.
#join='outer' → Mantém todas as colunas, adicionando valores NaN onde não houver correspondência.

# Verificando o resultado
df_total.head()# Mostrar as primeiras linhas do DataFrame final
```

## Depois de juntar os df, vamos filtrar os dados para poder organizar as informacoes
### 1 - Vamos pegar os valores unicos da coluna ['Nome UBS'] para poder filtrar ubs por ubs;
### 2 - Entao utilizamos um for para percorrer ubs por ubs e criar um grafico individualmente para cada uma;
### 3 - Dentro deste for vamos pegar pegar informacoes sobre os quadrimestres pq a intencao e criar um grafico mostrando o desenvolvimento de cada indicador nesse periodo, criando um grafico em que no eixo x tenha os indicadores, no eixo y o percentual e as colunas com os valores da quantidade dos 3 quadimestres por indicador

```python
import numpy as np

# UBSs únicas
ubs_list = df_total['Nome UBS'].unique()

# Criar gráficos para cada UBS
for ubs in ubs_list:
    df_ubs = df_total[df_total['Nome UBS'] == ubs]  # Filtra os dados da UBS específica

    # Posição das barras no eixo X
    x = np.arange(len(indicadores))

    # Definir o tamanho da figura
    plt.figure(figsize=(12, 6))

    # Criar as barras para os três quadrimestres
    quadrimestres = df_ubs['Quadrimestre'].unique()
    width = 0.2  # Largura das barras

    # Criar um for para pegar cada quadrimestre e plotar os 3 para cada indicador
    for i, quad in enumerate(quadrimestres):
        valores = df_ubs[df_ubs['Quadrimestre'] == quad][indicadores].values.flatten()
        bars = plt.bar(x + i * width, valores, width=width, label=quad)

        # Adicionar os valores acima das barras
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.0f}',
                     ha='center', va='bottom', fontsize=10, fontweight='bold')

    # Configurações do gráfico
    plt.xticks(x + width, indicadores, rotation=45, ha='right')  # Rotacionar os indicadores
    plt.xlabel('Indicadores')
    plt.ylabel('Percentual (%)')
    plt.title(f"Indicadores de Saúde - {ubs}")
    plt.legend(title="Quadrimestres")
    plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grade horizontal no fundo
    plt.gca().spines[['top', 'right']].set_visible(False)
    plt.tight_layout()

    # Mostrar o gráfico
    plt.show()
```
